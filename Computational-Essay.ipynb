### Introduction:
#### Where is intercountry adoption going?
Intercountry adoption has been a long-standing entrance way for families to grow and welcome new life into their home. By bringing families together, some across the country, some across oceans, intercountry adoption has been a pathway to loving families, especially for those who may not be able to have children at the time. In recent years, intercountry adoption years have been falling, with China as the leading country of the pack dropping from 8,000 annual adoptions to 1000 in a span of 15 years. This leaves the question of why? Is adoption still being recognized as an alternative to the family-lifestyle story? Is the decline due to an outside factor, hidden to the passing eye? The goal of this project was to analyze some of the potential factors leading to the rapid decline of intercountry adoptions worldwide. 

#### Who is involved?
Although it may seem like a subject that only affects a few particular families, the reality is that intercountry adoption affects a vast group of people. It acts as a way to spread culture and awareness and has gone onto strengthen relations amongst uncommon countries. Still at times an odd topic to discuss, you may be more affected by the colorful world of adoption and family growth than you think. 

#### Why should we care?
As a South Korean adoptee who grew up in rural Upstate New York, I have found that looking different amongst my family and friends has been my opportunity to share new cultures and outlooks with those around me. Adoption is widespread, it just doesn't always seem obvious. It is a beautiful way to introduce diversity into many lives and comes with beautiful experiences that adoptees get to share. At times, adoptees may feel like finding identity is a lot harder, due to the lack of specific culture they may have, but the reality is that being an adoptee is a culture within itself, and we get to share that with the world around us. Adoption is key to diversity, looking through the world in another lens amongst many. 

#### What are the factors involved?
Adoption is affected by many things: fluctuating populations, changing birthrates, a country's economic stability, policies constantly being enacted and then lifted. I worked towards analyzing some of these factors, looking at the top 5 countries involved in international adoption, along with their GDP over time and their involvement in international adoption groups. I also looked at the prices involved in the adoption process itself, as money can frequently push people away. There are many spots to look to for answers, and this is where I started.

#### What information was found from this data?
After digging and reformatting, I was able to find the top 5 countries involved in intercountry adoption's annuel adoptions per year over a 20 year time frame. I was shocked to see just how extreme the difference is between adoption numbers amongst China and the rest, with China maintaining a gap of at least 700 adoptions above the others. I was also able to collect the GDP data of these 5 countries over the span of 60 or so years, clearly illustrating the economic development of these countries.

### Results:
Below, you'll see a collection of scatter plots referencing the top 5 countries involved in intercountry adoption. The first collection displays the number of adoptions done over the course of 20 years, with a generally maintained pattern amongst the five being a decline, with the exception being Uganda. These countries were not displayed on the same set of axes due to China's clear difference in numbers, still maintaining adoption numbers in the thousands in 2019. The red line seen on the scatter plot for China's adoption numbers indicates when the one child policy was lifted as well. 

From there, you'll see the top 5 countries and their GDP patterns over the past 60 or so years. Most of them show an upward patterns, with an increasing GDP with minor fluctuations. Ukraine stands as the exception to this display. 

The display above clearly shows the falling numbers in intercountry adoptions, but it is hard to determine if this is purely due to economic stability of the countries involved, or if other factors are playing heavier roles than that can be seen. What is clear is that the intercountry adoption numbers are on a decrease, with fewer and fewer children being adopted every year.

One may argue that this isn't something that could be quantified, patterns change all the time and the decline isn't a negative or positive reflection of intercountry adoption. But I do worry that adoption isn't looked at a viable option for families enough. The thought behind this project was to bring light to a possible problem that should be discussed more, as we all deserve a loving and caring family, which some children do not get to experience.
